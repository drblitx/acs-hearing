```python
import pandas as pd
```

```python
df = pd.read_csv('../data/deaf_schools_2023/raw/k_12_deaf_2023.csv',index_col=False)
```

```python
df.head()
```

```python
df.columns
```

```python
# drop the enrol_nbr col (due to potentially outdated data)
df.drop(columns="enrol_nbr", inplace=True)

# drop the phone_nbr and vp_nbr col (due to reundant data; also needs verfification later on)
df.drop(columns=["phone_nbr", "vp_nbr"], inplace=True)
```

```python
# update the yrs_opr column to reflect additional years between 2023 and 2026
for index, row in df.iterrows():
    df.at[index, "yrs_opr"] = 2026 - row["yr_est"]

df.head()
```

```python
# export df to a clean csv
df.to_csv("../data/deaf_schools_2023/clean/deaf_schools2023.csv", index=False, na_rep="")
```
